```python
import scipy.stats as st
def norm_p(a, b):
    x = np.arange(-4, 4, 0.1)
    y = st.norm.pdf(x)
    x1 = x[(a <= x) & (x <= b)]
    y1 = y[(a <= x) & (x <= b)]
    p = st.norm.cdf(b) - st.norm.cdf(a)
    print(f"N(0,1) [{a},{b}] p={p}")
    plt.plot(x, y)
    plt.hlines(0,-4, 4)
    plt.vlines(x1,0,y1,colors='r')
norm_p(-2,2)
```

```python
from math import sqrt
def t_interval(x, b=0.95):
    a = 1 - b
    n = len(x)
    ta = st.t.ppf(1 -a/2, n-1)
    se = x.std()/sqrt(n)
    return x.mean() - ta*se, x.mean() + se*ta
t_interval(BSdata.身高)
```

```python
def ttest_1plot(X, mu=0, k=0.1):
    df = len(X) - 1
    t1p = st.ttest_1samp(X, popmean=mu)
    x = np.arange(-4,4,k)
    y = st.t.pdf(x, df)
    t = abs(t1p[0])
    p = t1p[1]
    x1 = x[x<=-t]; y1 = y[x<=-t]
    x2 = x[x>=t]; y2 = y[x>=t]
    print(f"样本均值:  {X.mean()}")
    print(f"单样本t检验: t={t}, p={p}")
    t_interval = st.t.interval(0.95,len(X)-1,X.mean(),X.std())
    print(f"置信区间: [{t_interval[0]},{t_interval[1]}]")
    plt.plot(x,y); plt.hlines(0,-4,4)
    plt.vlines(x1,0,y1,colors='r');plt.vlines(x2,0,y2,colors='r')
    plt.vlines(st.t.ppf(0.05/2,df),0,0.2,colors='b')
    plt.vlines(-st.t.ppf(0.05/2,df),0,0.2,colors='b')
    plt.text(-0.6,0.2,r"$\alpha$=%3.2f"%0.05,fontsize=15)
```

```python
ttest_1plot(BSdata.身高,166)
```

```python
ttest_1plot(BSdata.身高,170)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

population_mean = 1020
population_std = 100

sample_mean = 1080

z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(16))

alpha = 0.05
critical_value = st.norm.ppf(1 - alpha)
print("临界值:", critical_value)
if z_statistic > critical_value:
    print("拒绝")
else:
    print("不能拒绝")
```

```python
x = np.linspace(820, 1220, 1000)
y = st.norm.pdf(x, loc=population_mean, scale=population_std)
plt.plot(x, y, label='Original Distribution', color='blue')

# 填充拒绝区域
x_reject = np.linspace(population_mean + critical_value * population_std / np.sqrt(len(df)), 1200, 1000)
y_reject = st.norm.pdf(x_reject, loc=population_mean, scale=population_std)
plt.fill_between(x_reject, y_reject, color='red', alpha=0.3, label='Rejection Region')

# 标注样本均值和检验统计量
plt.axvline(x=sample_mean, color='green', linestyle='--', label='Sample Mean')
plt.text(sample_mean + 5, 0.002, 'Sample Mean', color='green')
plt.axvline(x=population_mean + critical_value * population_std / np.sqrt(len(df)),
            color='black', linestyle=':', label='Critical Value')
plt.text(population_mean + critical_value * population_std / np.sqrt(len(df)) + 5, 0.002,
         'Critical Value', color='black')

# 设置图形标题和坐标标签
plt.title('Hypothesis Testing')
plt.xlabel('Lifespan (hours)')
plt.ylabel('Probability Density')

# 图例和显示
plt.legend()
plt.show()
```
